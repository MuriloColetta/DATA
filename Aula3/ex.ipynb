```python
# Para rodar a aplicao
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configurao visual
sns.set_theme(style="darkgrid")

# Carregar base de dados
# Como o arquivo est na mesma pasta, basta colocar o nome
df = pd.read_csv('SampleSuperstore.csv')

# Mostrar as primeiras 5 linhas formatadas
display(df.head())

# Verificando tipos de dados e nulos
df.info()

# Anlise de Quantidade por Estado
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='State', y='Quantity', ci=None, palette='bright')
plt.title('Quantidade por Estado')
plt.xlabel('Estado')
plt.ylabel('Quantidade')
plt.xticks(rotation=90)
plt.show() 

# Comparao de Vendas por Segmento
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Segment', y='Sales', palette='Set2')
plt.title('Vendas por Segmento')
plt.xlabel('Segmento')
plt.ylabel('Vendas')
plt.show()

# Relao entre Desconto e Lucro
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Discount', y='Profit', hue='Category', palette='Set1')
plt.title('Relao entre Desconto e Lucro')
plt.xlabel('Desconto')
plt.ylabel('Lucro')
plt.show()

# Anlise de Vendas por Categoria
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Category', y='Postal Code', ci=None, palette='pastel')
plt.title('Vendas por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Vendas')
plt.show()

# Anlise de Lucro por Subcategoria
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Sub-Category', y='Profit', ci=None, palette='muted')
plt.title('Lucro por Subcategoria')
plt.xlabel('Subcategoria')
plt.ylabel('Lucro')
plt.show()
```
